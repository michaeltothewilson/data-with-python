### Environment Setup

Isolation is key. Always build your app in a virtual environment (`venv`) to avoid dependency hell.

__On Windows (PowerShell/CMD)__
1. Create: `python -m venv .venv`

2. Activate: `.\venv\Scripts\activate`

3. Install: `pip install streamlit pandas joblib scikit-learn`

__On Mac/Linux (Terminal)__
1. Create: `python3 -m venv .venv`

2. Activate: `source .venv/bin/activate`

3. Install: `pip3 install streamlit pandas joblib scikit-learn`

### Essential Streamlit Commands

* `st.title("Text")`: Creates a large `H1` header.

* `st.write("Text")`: Renders text, dataframes, and markdown.

* `st.sidebar`: Use as a prefix (`st.sidebar.slider()` for example) to move widgets to the left panel.

* `st.slider("Label", min, max, default)`: Captures numeric values (floats or integers).

* `st.button("Label")`: Returns True when clicked. Perfect for triggering predictions.

* `st.columns([1, 1])`: Splits the UI into side-by-side containers for better layout.

* `st.image("path.jpg", caption="...")`: Displays local images.

* `st.bar_chart(dataframe)`: A quick, interactive chart for showing species probabilities.

* `st.divider()`: Draws a horizontal line to separate logical sections of your app.

[Streamlit Official Cheat Sheet](https://docs.streamlit.io/develop/quick-reference/cheat-sheet)
