### 3.3 - Mapping Colors

Now, let's bring (some) order to chaos by mapping the values to standard colors.

In order to do this, we will create a `regex` map where the keys are the regular expressions and the value is the target color:

Here is a starter Map:

```python
regex_map = {
    r'^WH.*|^WHT|^WT|^W$': 'WHITE',
    r'^BK|^BLK|^B$': 'BLACK',
    r'^GR(AY|EY)|^GY|^GRY|^SIL.*|^SL$': 'GREY', 
    r'^BL$|^BLU': 'BLUE',
    r'^YEL.*|^Y[L|W]$': 'YELLOW',
    r'^BR.*|^BN$': 'BROWN',
    r'^RE.*|^RD$': 'RED',
}
```

### 3.4 - Some background

#### Understanding the Regex (The "Decoder Ring")
Regular Expressions act as a microscopic programming language dedicated entirely to pattern matching. To read the color map we just used, here is a breakdown of the specific symbols being used:

* `r'...'` (Raw String): Tells Python to treat everything inside the quotes as raw text, ignoring special escape characters.

* `^` (Starts With): An anchor that tells the engine to look only at the very beginning of the string. `^WH` means the text must start with `"WH"`.

* `$` (Ends With): An anchor for the end of the string. `^W$` means the string must start with `"W"` and immediately end, catching exactly `"W"` and nothing else.

* `|` (OR): Acts as a divider between different accepted patterns.

* `.*` (Wildcard): The dot (`.`) means "any single character" and the asterisk (`*`) means "zero or more times." Together, they mean "allow any characters after this point."

__Example Breakdown:__

The line `r'^WH.*|^WHT|^WT|^W$'` translates to:
Match the string if it starts with `WH` (and anything after), OR if it starts with `WHT`, OR if it starts with `WT`, OR if it is exactly the letter `W`.

#### Regex Traps

The color map looks like a good solution.

Take a moment to think critically about potential pitfalls regarding Regex:
* What might go wrong if we were to try to catch `"BLACK"` with `^B `?
* How would a `NaN` value affect the mapping process?
* Is a massive Regex dictionary even manageable?

#### Alternative Approaches to Consider
Depending on the dataset, there are other common methods for categorical cleaning:

##### The Manual Dictionary
Using the `.map()` method with a standard Python dictionary for explicit mapping.

__Example Code__:
```python
clean_map = {
    'WH': 'WHITE', 'WHT': 'WHITE', 'WT': 'WHITE', 'W': 'WHITE', 
    'BK': 'BLACK', 'BLK': 'BLACK', 'B': 'BLACK'
}

parking_df['Color'] = parking_df['Color'].map(clean_map)
```

##### Fuzzy Matching

For unstructured typos, packages like `thefuzz` use mathematical calculationa of how many single-character edits (insertions, deletions, or substitutions) it takes to change one string into another.

Example Code:

```python

messy_word = "WHIITTE"

# Returns the closest match and a confidence score
```
