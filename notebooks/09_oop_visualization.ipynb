```python
## Begin Imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
## End Imports
```

```python
file = "../data/opsd_germany_daily.csv"

df = pd.read_csv(file)

df["Date"] = pd.to_datetime(df["Date"])
df.set_index("Date", inplace=True)
df.info()
```

```python
## Focus on 2016
df_2016 = df[(df.index.year == 2016)]

df_2016.head()
```

```python
## Create Figure and Axes Object

fig, ax = plt.subplots(figsize=(10, 6))

# Plot Data
ax.plot(df_2016.index, 
        df_2016["Wind"],
        label="Wind Energy",
        color="lightgreen")

ax.plot(df_2016.index,
        df_2016["Solar"],
        label="Solar Energy",
        color="gold")

# Add Labels
ax.set_xlabel("Date")
ax.set_ylabel("Energy Production")
ax.set_title("Wind and Solar Energy Production Over Time")

ax.legend()

ax.grid(True, 
        linestyle="--", 
        alpha=0.7)

plt.tight_layout()
```
