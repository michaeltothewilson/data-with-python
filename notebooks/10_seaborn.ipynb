````python
# Create a Netflix Movie Dataframe
movie_mask = netflix_df["type"] == "Movie"
netflix_movies_df = netflix_df[movie_mask].copy()
netflix_movies_df.head()
````

````python
plt.figure(figsize=(5,6), 
           dpi = 100)
sns.boxplot(y="duration_minutes",
            data=netflix_movies_df)

plt.title("Distribution of Movie Duration (Minutes)")
plt.ylabel("Duration (Minutes)")
plt.tight_layout()
````

__Explanation__

`sns.boxplot(y=movies_df['duration'])`: This is the core of the box plot creation. 

We specify:
* `y=movies_df['duration']`: The numerical variable whose distribution we want to visualize is placed on the y-axis.
* If we wanted to compare across categories on the x-axis, we would use the `x` parameter.

Interpreting the Box Plot:

The resulting box plot will show you:

* The box: The middle 50% of the data falls within the box. The bottom of the box represents the 25th percentile (Q1), and the top represents the 75th percentile (Q3).
* The line inside the box: This line represents the median (the 50th percentile) of the movie durations.
* The whiskers: These lines extend from the box and typically represent the range of the data within 1.5 times the interquartile range (IQR = Q3 - Q1).
* The points outside the whiskers: These are potential outliers, data points that fall significantly outside the main distribution.

Now, let's say we wanted to compare the duration of movies across different content ratings (e.g., 'TV-PG', 'TV-MA', 'R'). We can use the x parameter in sns.boxplot() to specify the categorical variable:

````python
plt.figure(figsize=(12, 6), 
           dpi= 100)

sns.boxplot(data=netflix_movies_df,
            x = "rating",
            y = "duration_minutes")

plt.title("Movie Duration Distribution by Rating\n")
plt.xlabel("Rating")
plt.ylabel("Duration (Minutes)")
plt.tight_layout()
````

__Explanation__

`sns.boxplot(data=movies_df, x='rating', y='duration')`

* We use the `x` parameter to specify the `rating` column as the categorical variable that will be displayed on the x-axis

We use the `y` parameter for the numerical `duration` on the y-axis. This will create separate box plots for each unique rating within the movie dataframe.

This plot will allow you to visually compare the typical duration and the spread of movie durations for different maturity ratings. You might observe if certain ratings tend to have longer or shorter movies and if there are differences in the variability of their lengths.

## Scatter Plot - `sns.scatterplot()`

While seemingly basic, scatter plots are incredibly versatile for visualizing the relationship between two numerical variables. 

Seaborn enhances them by allowing you to incorporate additional dimensions using:

* `hue`: To represent a third categorical variable with different colors.
* `size`: To represent a third numerical variable with the size of the points.
* `style`: To represent a third categorical variable with different marker shapes.

Scatter plots are excellent for identifying correlations, clusters, and outliers in your data.

## Example

Using the titanic dataset, we can explore the relationship between `fare` and `age` with `survived` as the hue.

````python
## Begin Example
# Sample data
titanic = sns.load_dataset('titanic')

plt.figure(figsize=(10, 6))
sns.scatterplot(x='fare', y='age', hue='survived', data=titanic)
plt.title('Fare vs. Age (Colored by Survival)')
plt.xlabel('Fare')
plt.ylabel('Age')
plt.show()
## End Example
````

This scatter plot shows the relationship between the fare paid by passengers and their age. 

The `hue='survived'` argument colors the points based on whether the passenger survived. 

This allows us to visually inspect if there's any apparent correlation between these variables and survival. 

For instance, do we see more survivors at certain fare ranges or age groups?

## Histogram and Density Plot `sns.histplot()`

Histograms are crucial for understanding the distribution of a single numerical variable.

`sns.histplot()` shows the frequency of data points falling into different bins, giving you an idea of the data's shape, central tendency, and spread.

These plots are essential for identifying skewness, modality (number of peaks), and the overall shape of your data's distribution.

````python
# Histogram
plt.figure(figsize=(10, 6))
sns.histplot(titanic['age'], kde=True, bins=30)
plt.title('Distribution of Passenger Ages')
plt.xlabel('Age')
plt.ylabel('Frequency/Density')
````

This combines a histogram (showing the frequency of passengers in different age bins) with a Kernel Density Estimate (KDE) plot (a smoothed representation of the age distribution). 

This helps us understand the age range of passengers on the Titanic, identify any peaks or skewness in the distribution, and see the general shape of the age profile.

## Box Plot `sns.boxplot()`

Box plots are excellent for visualizing the distribution of a numerical variable across different categorical groups. 

They display:

* The median (the line inside the box).
* The quartiles (the edges of the box, representing the 25th and 75th percentiles).
* The interquartile range (IQR, the width of the box).
* Whiskers that extend to show the range of the data (typically 1.5 times the IQR).
* Outliers (individual points beyond the whiskers).
b
Box plots are fantastic for comparing the central tendency, spread, and presence of outliers across different categories.

## Example

Comparing `fare` distribution across different `pclass` (passenger class).


````python
## Begin Example
plt.figure(figsize=(10, 6))
sns.boxplot(x='pclass', y='fare', data=titanic)
plt.title('Fare Distribution by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Fare')
## End Example
````

This box plot shows how the distribution of fares paid differs across the three passenger classes (1st, 2nd, and 3rd). 

We can clearly see the median fare, the spread (IQR), and any outliers for each class. 

This visualization highlights the significant difference in fare prices based on passenger class.
