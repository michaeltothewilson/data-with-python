# Count Plot

Let's start with a simple visualization to understand the distribution of content types (Movies vs TV Shows) in our dataset.

We can use a countplot for this, which is essentially a bar chart that shows the frequency of each category.

Syntax:

```python
sns.countplot(data=df, x="value")
```

````python
sns.countplot(data=netflix_df,
              x="type")

plt.title("Distribution of Content Type\n")
plt.xlabel("\nContent Type")
plt.ylabel("Count\n")
plt.grid(axis="y", linestyle="--")
plt.tight_layout()
````

__Explanation__

* `sns.countplot(data=netflix_df, x='type')` creates the count plot. 

    * We specify the DataFrame (`netflix_df`) and the categorical variable we want to count (`type`) for the x-axis.

* `plt.title()`, `plt.xlabel()`, and `plt.ylabel()` are Matplotlib functions used to add labels and a title to our plot for better understanding.

This first plot gives us a quick visual representation of how many movies and TV shows are present in our dataset.

# Box Plots
Let's say we want to compare the distribution of content duration (in minutes for movies) for different content types (`Movie` vs. `TV Show`). 

Since the `duration` column likely represents different units for movies and TV shows, we should first filter our data to focus on movies where the `type` is `Movie`.

````python
# Create a Netflix Movie Dataframe
movie_mask = netflix_df["type"] == "Movie"
netflix_movies_df = netflix_df[movie_mask].copy()
netflix_movies_df.head()
````

````python
plt.figure(figsize=(5,6), 
           dpi = 100)
sns.boxplot(y="duration_minutes",
            data=netflix_movies_df)

plt.title("Distribution of Movie Duration (Minutes)")
plt.ylabel("Duration (Minutes)")
plt.tight_layout()
````

__Explanation__

`sns.boxplot(y=movies_df['duration'])`: This is the core of the box plot creation. 

We specify:
* `y=movies_df['duration']`: The numerical variable whose distribution we want to visualize is placed on the y-axis.
* If we wanted to compare across categories on the x-axis, we would use the `x` parameter.

Interpreting the Box Plot:

The resulting box plot will show you:

* The box: The middle 50% of the data falls within the box. The bottom of the box represents the 25th percentile (Q1), and the top represents the 75th percentile (Q3).
* The line inside the box: This line represents the median (the 50th percentile) of the movie durations.
* The whiskers: These lines extend from the box and typically represent the range of the data within 1.5 times the interquartile range (IQR = Q3 - Q1).
* The points outside the whiskers: These are potential outliers, data points that fall significantly outside the main distribution.

Now, let's say we wanted to compare the duration of movies across different content ratings (e.g., 'TV-PG', 'TV-MA', 'R'). We can use the x parameter in sns.boxplot() to specify the categorical variable:

````python
plt.figure(figsize=(12, 6), 
           dpi= 100)

sns.boxplot(data=netflix_movies_df,
            x = "rating",
            y = "duration_minutes")

plt.title("Movie Duration Distribution by Rating\n")
plt.xlabel("Rating")
plt.ylabel("Duration (Minutes)")
plt.tight_layout()
````

__Explanation__

`sns.boxplot(data=movies_df, x='rating', y='duration')`

* We use the `x` parameter to specify the `rating` column as the categorical variable that will be displayed on the x-axis

We use the `y` parameter for the numerical `duration` on the y-axis. This will create separate box plots for each unique rating within the movie dataframe.

This plot will allow you to visually compare the typical duration and the spread of movie durations for different maturity ratings. You might observe if certain ratings tend to have longer or shorter movies and if there are differences in the variability of their lengths.
